## Exploring Vector in Space

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vectors(vectors):
    """
    Plot a set of vectors, automatically labeled and annotated with coordinates.
    Automatically adjust plot limits based on vector values.
    """
    # Initialize the plot
    plt.figure(figsize=(6, 6))
    prop_cycle = plt.rcParams['axes.prop_cycle']
    colors = prop_cycle.by_key()['color']

    # Initialize the limits to the origin
    lims = np.array([0, 0, 0, 0])

    # Plot each vector
    for i, v in enumerate(vectors, start=1):
        color = colors[i % len(colors)]  # Cycle through colors
        plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color, label=f'Vector {i}')
        plt.text(v[0], v[1], f'v{i}={v}', fontsize=9, ha='right', va='bottom',fontweight='bold')  # Annotate coordinates

        # Update limits
        lims = np.vstack((lims, np.array([v[0]-1, v[0]+1, v[1]-1, v[1]+1])))

    # Set the plot limits
    plt.xlim(np.min(lims[:, 0]), np.max(lims[:, 1]))
    plt.ylim(np.min(lims[:, 2]), np.max(lims[:, 3]))

    # Add plot features
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend(fontsize="8")
    plt.title('2D Vectors Visualization')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Show the plot
    plt.show()

# Define any number of vectors
v1 = np.array([5, 3])
v2 = np.array([-5, -1])

# Plotting
plot_vectors([v1, v2])  # Automatically adjusts based on these vectors


```

## Sum and Difference of Vectors

```python
# Redefine the function to include annotation for the origin
def vector_addition(v1, v2):
    """
    Visualize the addition of two vectors v1 and v2 using the parallelogram method.
    Annotate the origin point as 'O'.

    Parameters:
    v1 (array-like): The first vector.
    v2 (array-like): The second vector.
    """
    # Create the vectors v1 and v2
    v1 = np.array(v1)
    v2 = np.array(v2)

    # Sum of vectors v1 and v2
    v3 = v1 + v2

    # Create a new figure and a set of subplots
    fig, ax = plt.subplots()

    # Plotting the vectors v1 and v2
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')
    ax.text(v1[0], v1[1], f'v1={v1}', fontsize=10, ha='right', va='bottom',fontweight='bold')  # Annotate coordinates
    ax.text(v2[0], v2[1], f'v2={v2}', fontsize=10, ha='left', va='bottom',fontweight='bold')  # Annotate coordinates

    # Plotting the resultant vector v3
    ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', label='v1 + v2')

    # Plotting the dashed lines to form the parallelogram
    ax.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', linestyle='dashed')
    ax.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', linestyle='dashed')

    # Annotate the origin 'O'
    ax.text(0, 0, 'O', color='black', fontsize=12, ha='right', va='top')

    # Setting the axes' limits to fit all vectors
    max_length = max(np.linalg.norm(v1), np.linalg.norm(v2), np.linalg.norm(v3))
    scaler=1.2
    ax.set_xlim(-max_length*scaler, max_length*scaler)
    ax.set_ylim(-max_length*scaler, max_length*scaler)

    # Adding labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title(f'v1 + v2 = {v3}')
    ax.legend(fontsize="8")

    # Showing the grid
    ax.grid(True)

    # Display the figure
    plt.show()

```

```python
# Define any number of vectors
v1 = np.array([3, 4])
v2 = np.array([5, 0])
# Call the function with example vectors
vector_addition(v1, v2)
```

# Vector Dot Product

![dot product](https://github.com/098765d/linear_algebra/blob/main/geo_dot_product.png?raw=true)

```python
def visualize_dot_product(v1, v2):
    """
    Visualize the dot product of vectors v1 and v2.
    Show the projection of one vector onto the other as a solid line.
    """
    # Calculate the dot product
    dot_product = np.dot(v1, v2)

    # Calculate the projection of v1 onto v2
    proj_of_v1_on_v2 = (np.dot(v1, v2) / np.linalg.norm(v2)**2) * v2

    # Start the plot
    fig, ax = plt.subplots(figsize=(6, 6))

    # Plot the original vectors as arrows
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')

    # Plot the projection of v1 onto v2 as a solid line
    ax.plot([0, proj_of_v1_on_v2[0]], [0, proj_of_v1_on_v2[1]], 'b-', label='Projection of v1 onto v2')

    # Draw the perpendicular line to v2 from the tip of v1 as a dashed line
    ax.plot([v1[0], proj_of_v1_on_v2[0]], [v1[1], proj_of_v1_on_v2[1]], 'purple', linestyle='dotted', label='Perpendicular to v2')

    # Annotate the vectors' coordinates
    ax.text(v1[0], v1[1], f'v1:{v1}', fontsize=11, ha='right')
    ax.text(v2[0], v2[1], f'v2:{v2}', fontsize=11, ha='left')

    # Annotate the origin 'O'
    ax.text(0, 0, 'O', color='black', fontsize=12, ha='right', va='top')

    # Annotate the angle theta
    # Code to calculate the angle theta between two vectors and annotate it on the plot
    theta = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

    # This line of code would annotate theta above the origin 'O'
    ax.text(0, 0, f'$\\theta = {np.degrees(theta):.0f}^\\circ$', fontsize=10, color='black', ha='center', va='bottom')


    # Set plot limits
    max_length = max(np.linalg.norm(v1), np.linalg.norm(v2), np.linalg.norm(proj_of_v1_on_v2)) + 2
    ax.set_xlim(-max_length, max_length)
    ax.set_ylim(-max_length, max_length)

    # Add plot features
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.title(f'Dot product of v1 and v2 = {dot_product}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Show the plot
    plt.show()

```

```python
# Define vectors
v1 = np.array([9, -3])
v2 = np.array([-5, -1])

# Visualize the dot product and the projection
visualize_dot_product(v1, v2)
```

```python
# Define vectors
v1 = np.array([3, 6])
v2 = np.array([7, 0])

# Visualize the dot product and the projection
visualize_dot_product(v1, v2)
```

```python

def plot_vector(v, label, color='blue'):
    """Plot a single vector with a label."""
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color,width=0.001)
    plt.text(v[0], v[1], label, color=color, fontsize=10, ha='right')

def visualize_vectors(vectors):
    """Visualize multiple vectors."""
    plt.figure(figsize=(8, 8))

    for label, vector in vectors.items():
        plot_vector(vector, label)

    # Determine plot limits
    all_values = np.array(list(vectors.values()))
    max_val = np.max(np.abs(all_values))

    plt.xlim(-max_val - 1, max_val + 1)
    plt.ylim(-max_val - 1, max_val + 1)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.title("Word Embeddings Visualization")
    plt.show()

# Define the word embeddings as a dictionary
word_embeddings = {
    'grandfather': np.array([-3.5, 4.375]),
    'man': np.array([-3.5, 2.375]),
    'adult': np.array([0.5, 2.375]),
    'woman': np.array([4.5, 2.375]),
    'boy': np.array([-3.5, -2.625]),
    'child': np.array([0.5, -2.625]),
    'girl': np.array([4.5, -2.625]),
    'infant': np.array([0.5, -3.625])
}


# Visualize the word vectors
visualize_vectors(word_embeddings)
```

## Text Similarity Using Normalized Dot Product

In this task, we explore the importance of normalized dot products (cosine similarity) in determining the similarity between words in natural language processing (NLP).

## Why Normalize Word Vectors?

Normalization is crucial in similarity calculations:
- **Balancing the Scale**: Without normalization, longer vectors (which might just have higher word frequencies) could dominate the similarity score, skewing the results.


## How to Normalize Word Vectors?

To normalize a word vector, we divide each element of the vector by its length. The length of a vector is calculated by taking the square root of the sum of the squares of its elements. For example, if we have a word vector `[3, 4]`, its length is $\sqrt{3^2 + 4^2} = 5$. To normalize it, we divide each element by 5, resulting in a normalized vector `[0.6, 0.8]`.




```python
# Make all the word embeddings unit vectors by normalizing them
for word, vector in word_embeddings.items():
    word_embeddings[word] = vector / np.linalg.norm(vector)
# Visualize the word vectors
visualize_vectors(word_embeddings)
```

```python
def calculate_similarity(word_embeddings):
    """Calculate the similarity (cosine similarity) between each pair of words."""
    words = list(word_embeddings.keys())
    similarities = {}

    for i in range(len(words)):
        for j in range(i + 1, len(words)):
            word1 = words[i]
            word2 = words[j]
            # Calculate dot product
            dot_product = np.dot(word_embeddings[word1], word_embeddings[word2])
            similarities[(word1, word2)] = dot_product
    similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    return similarities

# Calculate normalized similarities between word pairs
similarities = calculate_similarity(word_embeddings)
# Sort the dictionary by values in descending order
similarities
```
